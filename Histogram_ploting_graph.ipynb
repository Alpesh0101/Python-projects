# This is the Histogram plot.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import random
```

```python
Student_python=np.random.randint(20, 40, (100))
Student_machine=np.random.randint(15, 50, (100))
print(Student_python)
print(Student_machine)
```

```python
plt.hist(Student_python)
plt.show()
```

```python
plt.title("Python and Machine")
plt.xlabel("Student_python")
plt.ylabel("Student_machine")
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
bin=[20,25,30,35,40,50,60,80,90]
plt.figure(figsize=(20,8))
# plt.axis(0,50, 0,100)
Student_python=np.random.randint(20, 40, (100))
Student_machine=np.random.randint(50, 100, (100))
plt.hist([Student_python, Student_machine],bin, rwidth=0.8, histtype='bar', orientation='vertical', color=['b','g'], label=["python", 'machine'])
# plt.hist(Student_machine,bin, rwidth=0.6, histtype='bar', orientation='vertical', color='y', label="machine")
plt.legend()
# print(Student_python)
# print(Student_machine)
plt.grid(color='r', linestyle='--', linewidth='6')
plt.show()
plt.show()
```

```python
courses=['Python', 'DS', 'ML', 'DL']
student1=[19,25,30,41]
student2=[34,36,37,38]
student3=[23,2,5,24]
print(courses, student1)
```

```python
plt.figure(figsize=(8,6))
my_courses=np.arange(len(courses))
width=0.4
plt.bar(my_courses, student1, align='edge', color='y', edgecolor='r', linewidth=4, alpha=0.8, linestyle='--', visible=True)
plt.bar(my_courses, student2, align='center', color='g', edgecolor='b', linewidth=5, alpha=0.8, linestyle='--', visible=True)
plt.title('My Courses')
plt.xticks(my_courses+width, courses, rotation=30)
plt.xlabel('Courses', fontsize=18)
plt.ylabel('students', fontsize=20)
```

```python
plt.figure(figsize=(10,8))
my_courses=np.arange(len(courses))
width=0.4
plt.barh(my_courses, student1, align='edge', color='y', edgecolor='r', linewidth=4, alpha=0.8, linestyle='--', visible=True)
plt.barh(my_courses+width+width+width, student2, align='center', color='g', edgecolor='b', linewidth=5, alpha=0.8, linestyle='--', visible=True)
plt.title('My Courses')
plt.yticks(my_courses+width, courses, rotation=30)
plt.ylabel('Courses', fontsize=18)
plt.xlabel('students', fontsize=20)
```
