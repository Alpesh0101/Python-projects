```python
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
day=[11,22,23,24,25,40,33,39]
Mumbai=[12,14,15,26,27,31,45,49]
Panvel=[13,18,14,29,25,34,38,40]
style.use("ggplot")
plt.axis([0, 50, 0, 50])
plt.plot(day, Mumbai, "go--", linewidth=3, markersize=12, label='Mumbai line')
plt.plot(day, Panvel, "g^--", linewidth=3, markersize=12, label='Panvel line')
plt.title("Mumbai temperature", fontsize=20)
plt.xlabel('day', fontsize=15)
plt.ylabel('temperature', fontsize=15)
plt.legend(loc=2)
plt.grid(color='k', linestyle='--', linewidth=2)
plt.show()
```
